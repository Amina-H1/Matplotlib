```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
data_df = pd.merge(mouse_metadata, study_results, on='Mouse ID', how = "outer")
# Display the data table for preview
data_df
```

```python
# Checking the number of mice.
num_mice = len(data_df["Mouse ID"].value_counts())
print(f"The total number of mice is {num_mice}")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = data_df.loc[data_df.duplicated(subset = ["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
print(f"Duplicate Mouse ID:{duplicate_mice[0]}.")
```

```python
# Optional: Get all the data for the duplicate mouse ID.
data_df.loc[data_df["Mouse ID"] == "g989",:]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = data_df[data_df["Mouse ID"].isin(duplicate_mice) == False]
clean_data
```

```python
# Checking the number of mice in the clean DataFrame.
num_clean = clean_data["Mouse ID"].nunique()
print(f"The number of mice in the data is now {num_clean}")
```

```python
#Summary statistics
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
mean = clean_data.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
median = clean_data.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
variance = clean_data.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
standard_deviation = clean_data.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
sem = clean_data.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]

summary_df = pd.DataFrame({"Mean":mean, "Median":median, "Variance":variance, 
                               "Standard Deviation":standard_deviation, "SEM":sem})
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
aggregate_summary = clean_data.groupby("Drug Regimen").agg(["mean", "median", "var", "std", "sem"])["Tumor Volume (mm3)"]
aggregate_summary
```

```python
# BAR AND PIE CHARTS
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
plot_data = clean_data.groupby("Drug Regimen").count()["Timepoint"]
bar_plot = plot_data.plot.bar().set_ylabel("Number of Mice Tested")
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.
x_axis = clean_data["Drug Regimen"].unique()
y_axis = plot_data
plt.bar(x_axis, y_axis)
plt.xlabel("Drug Regimen",fontsize = 12)
plt.ylabel("Number of Mice Tested",fontsize = 12)
plt.title("Total Number of Timepoints for all Mice Tested")
plt.xticks(rotation = 90)
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
sex_plot = clean_data.groupby("Sex").count()["Mouse ID"]
pie_plot = sex_plot.plot.pie(autopct='%1.1f%%').set_ylabel("Sex")
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
labels = clean_data["Sex"].unique().tolist()
colors= ["royalblue", "lightcoral"]

plt.pie(sex_plot,labels = labels,colors=colors,autopct = "%1.1f%%")
plt.ylabel("Sex")
plt.title("Distribution of Male vs female Mice")
plt.show()
```

```python
#Quartiles, Outliers and Boxplots
# Calculate the final tumor volume of each mouse across four of the treatment regimens: 
# Capomulin, Ramicane, Infubinol, and Ceftamin
capomulin = clean_data.loc[clean_data["Drug Regimen"] == "Capomulin",:]
ramicane = clean_data.loc[clean_data["Drug Regimen"] == "Ramicane", :]
infubinol = clean_data.loc[clean_data["Drug Regimen"] == "Infubinol", :]
ceftamin = clean_data.loc[clean_data["Drug Regimen"] == "Ceftamin", :]

# Start by getting the last (greatest) timepoint for each mouse
capomulin_greatest = pd.DataFrame(capomulin.groupby('Mouse ID').max()['Timepoint'])
ramicane_greatest = pd.DataFrame(ramicane.groupby('Mouse ID').max()['Timepoint'])
infubinol_greatest = pd.DataFrame(infubinol.groupby('Mouse ID').max()['Timepoint'])
ceftamin_gretest = pd.DataFrame(ceftamin.groupby('Mouse ID').max()['Timepoint'])

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
capomulin_tvol = pd.merge(capomulin_greatest, data_df, on = ("Mouse ID", "Timepoint"), how = "left")
ramicane_tvol = pd.merge(ramicane_greatest, data_df, on = ("Mouse ID", "Timepoint"), how = "left")
infubinol_tvol = pd.merge(infubinol_greatest, data_df, on = ("Mouse ID", "Timepoint"), how = "left")
ceftamin_tvol = pd.merge(ceftamin_gretest, data_df, on = ("Mouse ID", "Timepoint"), how = "left")
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
box_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

# Locate the rows which contain mice on each drug and get the tumor volumes
capomulin_vol = capomulin_tvol["Tumor Volume (mm3)"]
ramicane_vol = ramicane_tvol["Tumor Volume (mm3)"]
infubinol_vol = infubinol_tvol["Tumor Volume (mm3)"]
ceftamin_vol = ceftamin_tvol["Tumor Volume (mm3)"]

box_data.append(capomulin_vol)
box_data.append(ramicane_vol)
box_data.append(infubinol_vol)
box_data.append(ceftamin_vol)

# add subset 
capomulin_quartiles = capomulin_vol.quantile([0.25, 0.5, 0.75])
capomulin_lower = capomulin_quartiles[0.25]
capomulin_upper = capomulin_quartiles[0.75]
capomulin_iqr = capomulin_upper = capomulin_lower

ramicane_quartiles = ramicane_vol.quantile([0.25, 0.5, 0.75])
ramicane_lower = ramicane_quartiles[0.25]
ramicane_upper = ramicane_quartiles[0.75]
ramicane_iqr = ramicane_upper = ramicane_lower

infubinol_quartiles = infubinol_vol.quantile([0.25, 0.5, 0.75])
infubinol_lower = infubinol_quartiles[0.25]
infubinol_upper = infubinol_quartiles[0.75]
infubinol_iqr = infubinol_upper = infubinol_lower

ceftamin_quartiles = ceftamin_vol.quantile([0.25, 0.5, 0.75])
ceftamin_lower = ceftamin_quartiles[0.25]
ceftamin_upper = ceftamin_quartiles[0.75]
ceftamin_iqr = ceftamin_upper = ceftamin_lower
    
# Determine outliers using upper and lower bounds
capomulin_lowlimit = capomulin_lower - (1.5 * capomulin_iqr)
capomulin_upplimit = capomulin_upper + (1.5 * capomulin_iqr)

ramicane_lowlimit = ramicane_lower - (1.5 * ramicane_iqr)
ramicane_upplimit =ramicane_upper + (1.5 * ramicane_iqr)

infubinol_lowlimit = infubinol_lower - (1.5 * infubinol_iqr)
infubinol_upplimit = infubinol_upper + (1.5 * infubinol_iqr)

ceftamin_lowlimit = ceftamin_lower - (1.5 * ceftamin_iqr)
ceftamin_upplimit = ceftamin_upper + (1.5 * ceftamin_iqr)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig, ax = plt.subplots()
ax.boxplot(box_data, labels = treatments)
ax.set_xlabel("Drug")
ax.set_ylabel("Tumor Volume (mm3)")
ax.set_title("Tumor Volume Boxplot")
plt.show()
```

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
line_plot = capomulin.loc[capomulin["Mouse ID"] == "l509"]
x_axis = line_plot["Timepoint"]
y_axis = line_plot["Tumor Volume (mm3)"]

plt.plot(x_axis, y_axis)
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin treatment of Mouse l509")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
scatter_plot = capomulin.groupby(["Mouse ID"]).mean()
x_axis = scatter_plot["Weight (g)"]
y_axis = scatter_plot["Tumor Volume (mm3)"]

plt.scatter(x_axis, y_axis)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse Weight vs. Volume (Averages)")
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = round(st.pearsonr(x_axis, y_axis)[0],2)
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_axis, y_axis)
regression = slope * x_axis + intercept
print(f"correlation: {correlation}")
plt.scatter(x_axis,y_axis)
plt.plot(x_axis,regression,"r-")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse Weight vs. Volume (Averages)")
plt.show()
```
